```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:\\Users\\soham\\Downloads\\householdtask3.csv')
```

```python
data.head(10)
```

```python
#scatter plot with year against own
plt.scatter(data['year'], data['own'])

#adding title to plot
plt.title("scatter plot")

#setting the x and y laels
plt.xlabel('year')
plt.ylabel('own')

#showing the result
plt.show()
```

```python
#line chart with year against own
plt.plot(data['year'])
plt.plot(data['own'])

#adding title to plot
plt.title("Line chart")

#setting the x and y laels
plt.xlabel('year')
plt.ylabel('own')

#showing the result
plt.show()
```

```python
#bar chart 
plt.bar(data['year'], data['own'])

#adding title to plot
plt.title("bar plot")

#setting the x and y laels
plt.xlabel('year')
plt.ylabel('own')

#showing the result
plt.show()
```

```python
#histogram
plt.hist(data['income'])

plt.title("histogram")

plt.show()
```
